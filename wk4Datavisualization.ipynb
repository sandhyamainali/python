## Compare GDP Growth of Nepal, India and China
- Bar Chart
- Line Graph

```python
from IPython.display import display
nepal_Ind_china_data=pd.read_csv('data/nep-india-china-gdp.csv')
# data load in pandas
display(nepal_Ind_china_data.head())
# plot pandas data
plt.figure(figsize=(15,9))

#to start fiure plt.figure
## define the width of the bars
width=0.5
x=range(len(nepal_Ind_china_data['Year']))
y=nepal_Ind_china_data['Nepal']
# range:x-axis segmrnt
# y:pandas data

# plotting the bars for nepal india and china
# y-axis ma katye segment vaney rw garda plt.bars
# width ali gap rakeu rw suru garna lai...- ani + width para samma 
plt.bar([i-width for i in x],nepal_Ind_china_data['Nepal'], label='Nepal', color='blue')
plt.bar(x,nepal_Ind_china_data['India'],label='India',color='red')
plt.bar([i+width for i in x],nepal_Ind_china_data['China'],label='China', color='yellow')

#adding title and labels
plt.title('GDP Growth Comparison: Nepal, India, China',fontsize=16)
plt.xlabel('Year',fontsize=14) #show x-axis name
plt.ylabel('GDP Growth(%)',fontsize=14) # show y-axis name
plt.xticks(x,nepal_Ind_china_data['Year'],fontsize=12, rotation=45) # show years name, value
plt.legend()

#adding grid line for better readability
plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.show()
```

## Line Graph

```python
import pandas as pd
import matplotlib.pyplot as plt
nepal_Ind_china_data=pd.read_csv('data/nep-india-china-gdp.csv')
years=nepal_Ind_china_data['Year']
nepal_gdp= nepal_Ind_china_data['Nepal']
india_gdp= nepal_Ind_china_data['India']
china_gdp= nepal_Ind_china_data['China']

# create plot
plt.figure(figsize=(20,10))
plt.plot(years,nepal_gdp, label='Nepal', color='Red',marker='.')
plt.plot(years,india_gdp, label='India', color='Blue',marker='x')
plt.plot(years,china_gdp, label='China', color='Green',marker='+')

plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.title('GDP Over Years for Nepal India and China')
# show legend color denote of particular area
plt.legend()
## ticks=minimun and maximum +1 mean value increment
plt.yticks(range(int(min(india_gdp)),int(max(china_gdp)) + 1, 1)) # yaxis with scale of 1
#if not tick 
plt.grid()
plt.show()
```

## Scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import numpy as np
nepal_economy=pd.read_csv('data/nepal-economy.csv')
# normalization using n/1000
nepal_economy['no_of_people_in_foreign_employment']= nepal_economy['no_of_people_in_foreign_employment'].apply(lambda n: n/1000)

display(nepal_economy.tail())
# create plot
plt.figure(figsize=(15,10))

plt.scatter(
    nepal_economy['GDP'],
    nepal_economy['no_of_people_in_foreign_employment'],
    color='purple',
    # edgecolors='black'
)
#nepal ecnomic foreign contrubution
# st linedraw
# if line draw in straight line then their will be co-relation
# to draw st line y=mx+c, but we don't have to use formula numpy already give this
plt.xlabel('GDP ')
plt.ylabel('Number of people in Foreign Employment')
plt.title('Number of Foreig Employment')
#plot ma line fit hunxa ki nai vaney rw herney
# def x, y to draw line
x= nepal_economy['GDP']
y= nepal_economy['no_of_people_in_foreign_employment']
m,c=np.polyfit( x , y , 1) # 1 is degree of polynoial y=mx+c is of degree 1
plt.plot( x , m * x + c, color='red', linewidth=1, label='GDP-FE-Trend-Line')
# show legend color denote of particular area
plt.legend()
## ticks=minimun and maximum +1 mean value increment
# plt.yticks(range(int(min(india_gdp)),int(max(china_gdp)) + 1, 1)) # yaxis with scale of 1
#if not tick 
# plt.xticks()
plt.grid()
plt.show()
```

## Bubble chart
- show relation between three relation

### HeatMap

```python
import seaborn as sns
plt.figure(figsize=(10,8))
## show correlation
correlation_matrix= nepal_economy.drop(columns=['Year'],inplace=False).corr()
#cmap:color
#annot= annot dekhauney ki na dekhauney 
#afai ley afai ko relation vayo vani..color warm hunxa
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Heatmap of correlation Matrix')
plt.show()
```

## Distribution
- Histogram
- Frequency Polygon
- Cumulative Frequency Distribution (ogive): Cumulative frequency distribution the facilitages to see how many observation lie above or below certain values is also graphical presenation of frequency
- Box Plot: It show the Five Point Summary (Min, Q1, Median, Q3)and outliers

```python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt

students_data=pd.DataFrame({
    'Marks':['1-20','20-40','40-60','60-80','80-100'],
    'no_of_students':[10,22,35,28,5]
})
students_data['mid_point']=students_data['Marks'].apply(lambda m: (int(m.split('-')[0]) + (int(m.split('-')[1]))/2))
lower_bounds=students_data['Marks'].apply(lambda m: int(m.split('-')[0])).min()
upper_bounds=students_data['Marks'].apply(lambda m: int(m.split('-')[1])).max()

bin_edge=list(range(lower_bounds,upper_bounds + 1,20))
students_data['bins']=pd.cut(x=students_data['no_of_students'],bins=bin_edge)
#binedge: histogram banauda kata bata suru garney
display(bin_edge)
display(students_data.head())

plt.figure(figsize=(10,5))
#use mid points as  the x-value for the histogram  and frequency y-values
plt.hist(
    students_data['mid_point'],
    bins=bin_edge,
    weight=students_data['no_of_students'],
    edgecolor='black',
    alpha=0.7,
    label='Marks'

)
plt.xlabel('Marks Range'('Mid point'))
```
