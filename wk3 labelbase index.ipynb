### Label Base Indexing(.loc[])
- label are columns, in python label are start by default with value 0
- We can customize lable by date, and so on
- .loc[] is used for selecting rows and column by labels
- We can use single labels, lists of label, or slice of labels
- can select row and column using loc[10:11,'name':'gender']
-[0:1] this are the location not indexing in pandas


```python
#selecting row based on level
import pandas as pd
from IPython.display import display
students=pd.read_csv('data\student-dataset.csv')
first_row=students.loc[0]
first_two_row=students.loc[0:10] # filter only row and give all column data
display(first_row) # row with label 1(index)label can be start
display(first_two_row)# row with labels 0 to 1

# Selecting specific rows to column
#if we need particular data of column and row, this is used 
first_five_stds_with_name_to_nationality= students.loc[0:4,'name':'nationality'] # selecting both row and column by using location (loc[0:1,'id','name])
display(first_five_stds_with_name_to_nationality)

first_and_second_stds_with_name_to_gender=students.loc[0:5,'name':'gender']
display(first_and_second_stds_with_name_to_gender)

data=students.loc[10:15,'city':'age']
display(data)
data1=students.loc[20:30,'english.grade':'language.grade']
display(data1)
```

## Integer location base indexing
- ilocal[] is used for selecting row and column
- iloc[]is indexingnot location
- giving indexing of row and column by iloc([0:1],[1,2])

```python
# integer base location indexing
display(students.iloc[1]) # second row(index 1)
display(students.iloc[0:3])

#selecting specific rows and column
display(students.iloc[0:2,1:3])# rows 0-1 and column 0-2, it will print value from 0-1 (:) is range
display(students.iloc[[0,2],[0,2]]) # rows 0-2 and column 0-2 it will give 0 and 2 value (,)is and 
```

-iat[]

```python
# Using .at[] to access single value by label
display(students.at[1,'name']) # value in row, column 'name
#label is used

#using .iat[] to access single value by position
#.iat indexing is used
display(students.iat[5,1]) # it will give 6th position, 5th indexing

display(students.iat[2,6])
```

### Logical Operator

```python
## Logical Operator
import pandas as pd
from IPython.display import display
students=pd.read_csv('data\student-dataset.csv')
display(students)

stds_fail_in_eng= students[students['english.grade']<2]
display(stds_fail_in_eng)
#stds with grade greater than 3.5
stds_pass_in_eng= students[students['english.grade']>3.5]
display(stds_pass_in_eng)

#select chinese std
stds_china=students[students['nationality'] == 'China']
display(stds_china)
topper_chinese_stds=stds_china[stds_china['english.grade']> 3.6]
display(topper_chinese_stds)
```

### Multiple Logical Operator

```python
from IPython.display import display
import pandas as pd
students=pd.read_csv('data\student-dataset.csv')
#students having grades greater then 3.5 in all subjects
tooper_students= students[( students ['math.grade'] > 3.6) & (students['english.grade'] > 3.6)]
display(tooper_students)
```

```python
# OR operator
from IPython.display import display
import pandas as pd
students=pd.read_csv('data\student-dataset.csv')
#students having grades greater then 3.5 in all subjects
Stds_high_mark= students[( students ['math.grade'] > 3.8) | (students['english.grade'] > 3.8)]
display(Stds_high_mark)
```

```python
# find all indian female Students
import pandas as pd
from IPython.display import display
students=pd.read_csv('data\student-dataset.csv')
# to filter data we have to use value and name of column to find value of data
female_stds= students[ (students ['gender'] == 'F') & (students['nationality'] == 'India') & (students['age'] >= 23) & (students['math.grade'] >= 3.8)]
female_stds
```

### filter data using
- str
- Isin
- contains
- between
- startWith

```python
import pandas as pd
from IPython.display import display
#startWith is use to find name startwith j or whatever we want
students= pd.read_csv('data\student-dataset.csv')
std_with_j= students[students.name.str.startswith('J')]
std_with_j.head()
# contains is used to see, if their is any name contains with assign value or not
stds_with_strunk=students[students.name.str.contains('Strunk')]
stds_with_strunk

#students with math.grade beth 4 and 5 and use in either number or date
stds_with_name_j=students[students['math.grade'].between(4,5)]
stds_with_name_j.head()
#chines or anerican 
#while using isin should use small bracket with []
chines_or_american= students[students['nationality'].isin(['china','American'])]
chines_or_american

#student with age betn 20-24 
#while using between ,start with small bracket
std_age=students[students['age'].between(20,24)]
std_age
```

## Tilde (~)
- not equal to

```python
# Filter students with the age less than 20 and greater than 24
std=students[~students['age'].isin([20,30])]
std.head()
```

## Query

```python
# Filter std with age above 25 and gender is 'male
query_filter=students.query("age > 25 and gender == 'M'")
query_filter
```

## nlargest and nsmallest
- use nlargest and nsmallest to get rows with the top or bottom values in column
- nlargest-top values
- nsmallest-bottom values


```python
# Get top 3 students with the highest grades in english
top_3_marks=students.nlargest(3,'english.grade')
top_3_marks

#get 2 students with the lowest grade in english
botton_2_marks=students.nsmallest(2,'english.grade')
botton_2_marks
```
