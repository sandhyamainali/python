# Handle missing data
- Remove Rows Containing Missing values (df.dropna()) -millinon data (drop)
    - dropna(axis=0) removes rows containing at least one missing value
- Replace Missing Value (df.fillna()) -small data(fill)

```python
from IPython.display import display
import pandas as pd
students= pd.read_csv('data\student-dataset.csv')
display(students.head())
students.dropna(axis=0,how='any',inplace=True) # remove all row with value NaN
# (axis = 0 default) - drop with NaN
students=students.fillna(1) #replace missing value with 1
display(students.head())
students.fillna(method='ffill') # fill with previous row value
students.fillna(method='bfill')# fill with next row
students['english.grade'].fillna(50) #only fill value of english grade
students.fillna({'english.grade':50,'math.grade':60})#replace english with math
df_interpolated=students.interpolate()
#interpolates: value fill using interpolate
students.fillna(df_interpolated)
```
