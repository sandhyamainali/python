## Grouping & Aggregrate
- grouping consist of three operations
    - split
    - Apply
    - Combine

```python
import pandas as pd
students= pd.read_csv('data\student-dataset.csv')
#group data based on gende
gender_data= students.groupby('gender')
display(gender_data.count())

gender_data['english.grade'].mean()

display(gender_data['math.grade'].mean())
group_nationality=students.groupby('nationality')
display(group_nationality['math.grade'].mean())
```
