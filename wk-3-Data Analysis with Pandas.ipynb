### can index row and column
- column can use name inplace of number
- like (2:3),(Age:Address)
- index is used for selecting data

```python
# Basic indexing
import pandas as pd
from IPython.display import display
student=pd.read_csv('data\student-dataset.csv')
display(student['name']) # single column as a series
new_std_data=student[['id','name','nationality']]
display(new_std_data) #multiple column dataframe

## slicing row
first_10_rows=student[0:10]
display(first_10_rows) #only first row dataframe
last_10_rows=student[302:201].copy()

display(last_10_rows) # row from index 30 to last index (dataframe)

#first 5 rows with only name and nationality
student[0:5][['name','nationality']]

```
