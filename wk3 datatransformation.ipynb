## Data Transformation and Mapping
- apply:allow us to apply a function to each element row, or column in a DataFrame or series
- map:is used to map values in series according to dictionary or another series
- replace:replace allow for replacing specific values in Dataframe with new values
- astype: this is used to convert data type
- pipe:it allows chaining and using complex function that operat on the entire DataFrame. This can be used for complex transformation operation

## apply
- simple comlumn value change

```python
import pandas as pd
import numpy as np
#convert grade to scale of 100
students= pd.read_csv('data\student-dataset.csv')
students['maks_in_english'] = students ['english.grade'].apply(lambda grade: grade * 10)
    
students.head()
```

### Map
- value replace with new category

```python
import pandas as pd
students=pd.read_csv('data\student-dataset.csv')
students['gender']=students['gender'].map({'M':'Male','F':'Female'})
students['nationality']=students['nationality'].map({'China':'Chinese','United States of America':'American', 'Brazil': 'Japan'})
students.head()
```

## replace

```python

#replace name John Doe with jayram (replace specific value)
students['name']=students['name'].replace({'Kiana Lor':'JayRam'})
students.head()
```

### Astype
- convert value

```python
#convert english.grade column to int
students['english.grade']=students['english.grade'].astype(int)
students['math.grade']=students['math.grade'].astype(int)
students.head()
```

## pipe
- complex cal in particular column

```python
def calc(df):
    df['final_grades']= df['english.grade'].astype(float)+ df['math.grade'].astype(float)
    return df
students=students.pipe(calc)
students.tail()
```
