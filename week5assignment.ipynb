## **Predictive Analytics using Machine Learning**
## Summary of Findings
The main objective of the regression models involved forecasting both inflation data and Consumer Price Index (CPI) values.  Among the studied models were linear regression in combination with decision- tree regression and support vector regression (SVR) and random forest regression.  Model assessment employed the mean squared error (MSE), Root Mean Squared Error (RMSE) and R2 Score as evaluation metrics.  The error power evaluation provided by RMSE determines the square root of that error in contrast to the MSE capability which calculates mean square differences between observed and predicted values.  A model fits better when the R2 Score improves because it indicates better variable explanation of the target variable.
Linear regression acted as a starting model to identify basic variable correlations however it showed limited effectiveness when dealing with complex data patterns.  With its built-in capability to split data through important branches the Decision Tree Regressor can detect non-linear patterns within its significant variables.  The model shows evidence of workout set overfitting when its adjustment parameters are incorrect.  Random Forest Regressor increases accuracy and generalizability along with reducing overfitting through the averaging of multiple decision tree forecasts.  The purpose of SVR is to create the optimal hyperplane match for data sets but this approach requires challenging parameter adjustments for successful functionality (Terra, 2024)

Model accuracy depends on both high R2 values and low values of MSE and RMSE for accurate performance assessment.  The data contained complex non-linear correlations and the decision tree together with random forest regressions demonstrated the most efficient detection capability.  The simulated complexity which linear regression models failed to reproduce because of its simpler nature. Various SVR parameter adjustments allowed it to reproduce non-linear patterns effectively.

 The performance assessment of models occurred through visual comparison in scatter plots between actual and predicted values.  Linear Regression revealed inaccurate demands as it yielded large variances in its measurements yet Random Forest and other models produced closely corresponding predictions between actual results.





```python
import pandas as pd
import numpy as np
from functools import reduce
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.svm import SVR, SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, precision_recall_fscore_support, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Load data from each file
remittance_path = "data/Remittance_Data.csv"
trade_path = "data/Trade_Data.xlsx"
interest_rates_path = "data/Interest_Rates_Data.xlsx"
gdp_growth_path = "data/GDP_Growth_Data.xlsx"
cpi_inflation_path = "data/CPI_Inflation_Data.xlsx"

# Load data
remittance_data = pd.read_csv(remittance_path)[["Year", "Remittance Inflow (Rs. billion)", "Growth (%)"]]
trade_data = pd.read_excel(trade_path, engine='openpyxl')[["Year", "Exports (Rs. billion)", "Imports (Rs. billion)"]]
interest_rates_data = pd.read_excel(interest_rates_path, engine='openpyxl')[["Year", "Repo Rate (%)", "Reverse Repo Rate (%)"]]
gdp_growth_data = pd.read_excel(gdp_growth_path, engine='openpyxl')[["Year", "GDP Growth (%)"]]
cpi_inflation_data = pd.read_excel(cpi_inflation_path, engine='openpyxl')[["Year", "CPI Inflation (%)"]]

# Merge all datasets on 'Year'
data_frames = [remittance_data, trade_data, interest_rates_data, gdp_growth_data, cpi_inflation_data]
consolidated_data = reduce(lambda left, right: pd.merge(left, right, on='Year'), data_frames)
display(consolidated_data)
# Calculate trade deficit
consolidated_data['Trade Deficit (Rs. billion)'] = consolidated_data['Imports (Rs. billion)'] - consolidated_data['Exports (Rs. billion)']

# Feature selection and target variable for regression
features = consolidated_data.drop(columns=["Year", "CPI Inflation (%)"])
target_regression = consolidated_data["CPI Inflation (%)"]
display(features)


# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
display(scaled_features)
# Train-test split for regression
X_train, X_test, y_train_regression, y_test_regression = train_test_split(scaled_features, target_regression, test_size=0.2, random_state=42)
display("Train and Test Split", X_test, X_train)

# Define regression models
regression_models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree Regressor": DecisionTreeRegressor(random_state=42),
    "Random Forest Regressor": RandomForestRegressor(random_state=42),
    "SVR": SVR()
}

# Train and evaluate regression models
regression_results = {}
for name, model in regression_models.items():
    model.fit(X_train, y_train_regression)
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test_regression, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test_regression, predictions)
    regression_results[name] = {"MSE": mse, "RMSE": rmse, "R² Score": r2}

# Show regression results
display("\nRegression Model Evaluation Results:")
for model_name, metrics in regression_results.items():
    display(f"{model_name}: MSE = {metrics['MSE']:.4f}, RMSE = {metrics['RMSE']:.4f}, R² Score = {metrics['R² Score']:.4f}")

# Plot actual vs predicted values for regression models
plt.figure(figsize=(12, 6))
plt.scatter(y_test_regression, predictions, color='blue', label="Predictions")
plt.plot([min(y_test_regression), max(y_test_regression)], [min(y_test_regression), max(y_test_regression)], color='red', linestyle='--')
plt.title('Actual vs Predicted Values (Regression)')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.legend()
plt.show()


# Define target for classification (1 = growth, 0 = decline)
classification_target = (consolidated_data["GDP Growth (%)"] > 0).astype(int)

# Train-test split for classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(scaled_features, classification_target, test_size=0.2, random_state=42)



# Classification: Predicting Economic Recession (1 = Recovery, 0 = Recession)
classification_target = (consolidated_data["GDP Growth (%)"] > 0).astype(int)  # 1 for recovery, 0 for recession

# Train-test split for classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(scaled_features, classification_target, test_size=0.2, random_state=42)

# Define classification models
classification_models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier(random_state=42),
    "Random Forest Classifier": RandomForestClassifier(random_state=42),
    "SVM": SVC(probability=True),  # Enable probability estimation
    "KNN": KNeighborsClassifier()
}

# Train and evaluate classification models
classification_results = {}
for name, model in classification_models.items():
    model.fit(X_train_class, y_train_class)
    predictions = model.predict(X_test_class)
    accuracy = accuracy_score(y_test_class, predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(y_test_class, predictions, average='binary')
    conf_matrix = confusion_matrix(y_test_class, predictions)
    fpr, tpr, thresholds = roc_curve(y_test_class, model.predict_proba(X_test_class)[:, 1])
    auc_score = auc(fpr, tpr)
    
    classification_results[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "Confusion Matrix": conf_matrix,
        "AUC Score": auc_score
    }

# Show classification results

for model_name, metrics in classification_results.items():
    display('Classification Model Evaluation Results:',f"{model_name}: Accuracy = {metrics['Accuracy']:.4f}, Precision = {metrics['Precision']:.4f}, Recall = {metrics['Recall']:.4f}, F1 Score = {metrics['F1 Score']:.4f}, AUC Score = {metrics['AUC Score']:.4f}")
    
    # Plot Confusion Matrix
    sns.heatmap(metrics["Confusion Matrix"], annot=True, fmt="d", cmap='Blues', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Plot ROC Curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {metrics["AUC Score"]:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.title(f'ROC Curve - {model_name}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.show()

# Show classification results

for model_name, metrics in classification_results.items():
    display('Classification Model Evaluation Results:',f"{model_name}: Accuracy = {metrics['Accuracy']:.4f}, Precision = {metrics['Precision']:.4f}, Recall = {metrics['Recall']:.4f}, F1 Score = {metrics['F1 Score']:.4f}, AUC Score = {metrics['AUC Score']:.4f}")
    
    # Plot Confusion Matrix
    sns.heatmap(metrics["Confusion Matrix"], annot=True, fmt="d", cmap='Blues', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Plot ROC Curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {metrics["AUC Score"]:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.title(f'ROC Curve - {model_name}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.show()
```
