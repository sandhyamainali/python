## Confusion Matrix
- indentifie true idnetified. (If condn os true output must be true, if false output should be false)
- A confusion Matrix is a performance evaluation tools in machine learning, representation the accuracy of a classification model.
- It display the number of true position, true negative, false positive,, false negative . This matrix aids in analyzing model performance, identifying mis-classification, and improving predicting accuracy.
- True Positive- 560, meaning the model correctly classified 560 positive class data points.
- True Negative- 330, meaning the model correctly classified 330 negative class data points
- False POsitive- 60,meaning the model incorrectly classified negative class data points as belonging to the positive class
- False Negative- 50, meaning the model incorrectly classified 50 positive class data points to the negative class

- Accuracy:TP+TN/TP+FN+TN+FN
- Precision: TP/TP+FP
- Recall:TP/TP+FN

## K-Means clusturing Algorithm
- choose Number of cluster (k): We decide how amny cluster you want the data divided into.
- Initialize cluster center: Randomly select k points from the data as initial cluster centers (also called centroids).

```python
import pandas as pd
home_data = pd.read_csv('data/housing.csv', usecols = ['longitude', 'latitude', 'median_house_value'])
home_data.head()
```

```python
## plot scattered plot to visualization the data
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,4))
sns.scatterplot(data=home_data,x='longitude',y='latitude',hue='median_house_value')
plt.show()
```

```python
## Split Tain/Test and Normalization Data

from sklearn.model_selection import train_test_split
from IPython.display import display
from sklearn import preprocessing
X_train,X_test, y_train, y_test= train_test_split(home_data[['latitude','longitude']],home_data[['median_house_value']],test_size=0.33,random_state=0)
X_train_norm=preprocessing.normalize(X_train)
X_test_norm=preprocessing.normalize(X_test)

display(X_train_norm)

display(X_test_norm)
```

```python
from sklearn.cluster import KMeans
Kmeans= KMeans(n_clusters=3, random_state = 0, n_init='auto')
Kmeans.fit(X_train_norm)

plt.figure(figsize=(10,5))
sns.scatterplot(data=X_train,x='longitude', y='latitude',hue=Kmeans.labels_)
plt.show()
```
